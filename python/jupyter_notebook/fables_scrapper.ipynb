```python
from bs4 import BeautifulSoup
import requests
import urllib3
import json
```

```python
urllib3.disable_warnings(
    urllib3.exceptions.InsecureRequestWarning
)
```

```python
aesop =["aesop01","aesop02"]
```

```python
url ="https://learnsanskrit.cc/fables/story?name=aesop02&active=true"
```

```python
page = requests.get(
    url,
    verify=False
)
```

```python
complete_data = page.json()
```

```python
print(json.dumps(complete_data, indent=4))
```

```python
title_data= complete_data['data']['summary_head']
```

```python
print(title_data)
```

```python
title = title_data[0]
```

```python
# Clean text
title = title.replace("The ", "")
title = title.replace("the ", "")
title = title.replace(" and ", ", ")

# Split into separate animal names
actors = title.split(", ")
```

```python
print(actors)
```

```python
story_moral = title_data[1]
```

```python
print(story_moral)
```

```python
story_moral = story_moral.replace("(","")
story_moral = story_moral.replace(")","")
print (story_moral)
```

```python
english_version = complete_data['data']['summary_text']
```

```python
print(english_version)
```

```python
transliterated_version = []

transliterated_texts = complete_data['data']['texts']

for section in transliterated_texts:

    soup = BeautifulSoup(section, "html.parser")

    divs = soup.find_all("div")

    words = []

    for div in divs:

        data = div.get_text(strip=True)

        # Remove spaces/newlines
        data = data.replace("\n", "").replace(" ", "")

        # Skip empty values
        if data == "":
            continue

        # Skip numbers
        if data.isdigit():
            continue

        words.append(data)

    # Convert block into one sentence/string
    if words:
        transliterated_version.append(" ".join(words))

print(transliterated_version)
```

```python
sanskrit_version = []

sanskrit_texts = complete_data['data']['textsdeva']

for section in sanskrit_texts:

    soup = BeautifulSoup(section, "html.parser")

    divs = soup.find_all("div")

    words = []

    for div in divs:

        data = div.get_text(strip=True)

        # Remove spaces/newlines
        data = data.replace("\n", "").replace(" ", "")
        data = data.replace("\\","").replace(" ","")

        # Skip empty values
        if data == "":
            continue

        # Skip numbers
        if data.isdigit():
            continue

        words.append(data)

    # Convert block into one sentence/string
    if words:
        sanskrit_version.append(" ".join(words))

print(sanskrit_version)
```

```python
data = {
    "title": title,
    "actors": actors,
    "storyMoral": story_moral,
    "englishVersion": english_version,
    "transliteratedVersion": transliterated_version,
    "sanskritVersion": sanskrit_version
}

print(data)
```

```python
with open("stories.json", "a", encoding="utf-8") as f:
    json.dump(
        data,
        f,
        ensure_ascii=False,
        indent=4
    )
```
