```python
from bs4 import BeautifulSoup
import requests
import urllib3
import json

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

url = "https://learnsanskrit.cc/fables/getFables"

page = requests.get(url, verify=False)

# Parse HTML instead of trying to read JSON
soup = BeautifulSoup(page.text, "html.parser")

# Example: print formatted HTML
print(soup.prettify())

# If you want structured extraction instead:
data = {
    "title": soup.title.string if soup.title else None
}

print(json.dumps(data, indent=4))
```

```python
img_tags = soup.find_all("img")
```

```python
print (img_tags)
```

```python
jpg_names = []

for img in img_tags:
    src = img.get("src")

    if src:
        # clean bad characters
        src = src.replace('\\"', '').replace('"', '').strip().strip(",")

        # get filename only
        filename = src.split("/")[-1]

        if filename.lower().endswith(".jpg"):
            jpg_names.append(filename)

print(jpg_names)
```

```python
cleaned_fable_names = [name.replace("_icon.jpg","") for name in jpg_names]
```

```python
print(cleaned_fable_names)
```

```python
structured = {
    "aesop": [],
    "pancha": [],
    "hito": [],
    "jataka": []
}

for name in cleaned_fable_names:
    for key in structured.keys():
        if key in name:
            structured[key].append({
                "storyNumber": name,
                "used": False,
                "updatedOn": None
            })

print(structured)
```
