```python
import os
import json
import stanza
```

```python
nlp = stanza.Pipeline("sa")
```

```python
file_path = os.path.join("..","data","stories.json")
print(file_path)
```

```python
stories =[]
```

```python
with open(file_path, "r", encoding="utf-8") as f:
    stories = json.load(f)

print(stories) 
```

```python
sanskrit_version_story = stories["sanskritVersion"]
print(sanskrit_version_story)
```

```python
tokenized_sanskrit = []

for sentence in sanskrit_version_story:

    doc = nlp(sentence)

    for sent in doc.sentences:
        for word in sent.words:
            print(
                f"word: {word.text}\t"
                f"upos: {word.upos}\t"
                f"xpos: {word.xpos}\t"
                f"feats: {word.feats if word.feats else '_'}"
            )
```

```python
tokenized_sanskrit = []

for sentence in sanskrit_version_story:

    doc = nlp(sentence)

    sentence_tokens = []

    for sent in doc.sentences:
        for word in sent.words:
            sentence_tokens.append({
                "text": word.text,
                "upos": word.upos,
                "xpos": word.xpos,
                "feats": word.feats if word.feats else "_"
            })

    tokenized_sanskrit.append(sentence_tokens)
```

```python
story["tokenized_sanskrit_version"] = tokenized_sanskrit

with open(file_path, "w", encoding="utf-8") as f:
    json.dump(story, f, indent=4, ensure_ascii=False)

print("Tokenized Sanskrit data saved successfully!")
```
