```python
import json

results = []

for entry in kosha.get("kAkasya"):
    data = {
        "gender": str(entry.linga),
        "case": str(entry.vibhakti),
        "number": str(entry.vacana),
    }

    pratipadika_entry = entry.pratipadika_entry

    if hasattr(pratipadika_entry, "pratipadika"):
        data["lemma"] = pratipadika_entry.pratipadika.text
        data["analysis_type"] = "basic"

    results.append(data)

print(json.dumps(results, indent=2, ensure_ascii=False))
```

```python
import json

results = []
seen = set()

for entry in kosha.get("kAkasya"):
    pratipadika_entry = entry.pratipadika_entry

    # Keep only basic analyses
    if not hasattr(pratipadika_entry, "pratipadika"):
        continue

    lemma = pratipadika_entry.pratipadika.text

    data = {
        "lemma": lemma,
        "analysis_type": "basic",
        "gender": str(entry.linga),
        "case": str(entry.vibhakti),
        "number": str(entry.vacana),
    }

    key = (
        lemma,
        data["gender"],
        data["case"],
        data["number"],
        data["analysis_type"]
    )

    if key not in seen:
        seen.add(key)
        results.append(data)

print(json.dumps(results, indent=2, ensure_ascii=False))
```

```python
import sys
from vidyut.kosha import Kosha, PadaEntry


def get_all(kosha: Kosha, query: str) -> list:
    if query.endswith("H"):
        queries = [query[:-1] + 's', query[:-1] + 'r']
    else:
        queries = [query]

    results = []
    for query in queries:
        results += kosha.get(query)
        return results


def display_entry(e: PadaEntry):
    match e:
        case PadaEntry.Subanta():
            if e.is_avyaya:
                print(f"avyaya: {e.lemma:10}")
            else:
                print(f"subanta: {e.lemma:10} {e.linga}, {e.vibhakti}, {e.vacana}")
        case PadaEntry.Tinanta():
            print(f"tinanta: {e.lemma:10} {e.prayoga}, {e.lakara}, {e.purusha},  {e.vacana}")
```

```python
entries = get_all(kosha,"kAkasya")
for entry in entries:
    display_entry(entry)
```

```python
from vidyut.kosha import Kosha, PadaEntry
from vidyut.lipi import Scheme, transliterate, detect


def display_entry(e: PadaEntry, output_scheme: Scheme):
    match e:
        case PadaEntry.Subanta():
            if e.is_avyaya:
                print(f"avyaya: {e.lemma}")
            else:
                print(
                    f"subanta: {e.lemma} "
                    f"{e.linga}, {e.vibhakti}, {e.vacana}"
                )

        case PadaEntry.Tinanta():
            print(
                f"tinanta: {e.lemma} "
                f"{e.prayoga}, {e.lakara}, "
                f"{e.purusha}, {e.vacana}"
            )


def run(query: str, output_scheme: Scheme = Scheme):
    # Detect input scheme and convert to SLP1
    encoding = detect(query) or Scheme.HarvardKyoto
    slp_query = transliterate(query, encoding, Scheme.Slp1)

    kosha = Kosha("vidyut-0.4.0/kosha")

    for entry in kosha.get(slp_query):
        display_entry(entry, output_scheme)


run("AsIt",Scheme.Iast)
```

```python
from vidyut.prakriya import Vyakarana, Pada, Dhatu, Prayoga, Lakara, Purusha, Vacana


def show_tinantas(dhatu: Dhatu, output_scheme: Scheme):
    v = Vyakarana()
    for lakara in Lakara.choices():
        pada = Pada.Tinanta(dhatu, Prayoga.Kartari, lakara, Purusha.Prathama, Vacana.Eka)
        prakriyas = v.derive(pada)
        for p in prakriyas:
            text = transliterate(p.text, Scheme.Slp1, output_scheme)
            print(f"- {text}")


def display_entry(e: PadaEntry, output_scheme: Scheme):
    lemma = transliterate(e.lemma, Scheme.Slp1, output_scheme)
    match e:
        case PadaEntry.Subanta():
            if e.is_avyaya:
                print(f"avyaya: {e.lemma:10}")
            else:
                print(f"subanta: {e.lemma:10} {e.linga}, {e.vibhakti}, {e.vacana}")
        case PadaEntry.Tinanta():
            print(f"tinanta: {e.lemma:10} {e.prayoga}, {e.lakara}, {e.purusha}, {e.vacana}")
            show_tinantas(e.dhatu_entry.dhatu, output_scheme)
```
