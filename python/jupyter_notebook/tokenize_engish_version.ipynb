```python
import os
import json
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
file_path = os.path.join("..","data","stories.json")
print(file_path)
```

```python
stories =[]
```

```python
with open(file_path, "r", encoding="utf-8") as f:
    stories = json.load(f)

print(stories)    
```

```python
english_version_story = stories["englishVersion"]
print(english_version_story)
```

```python
doc = nlp(english_version_story)
```

```python
tagged_words = []

for token in doc:
    tagged_words.append({
        "text": token.text,
        # "lemma": token.lemma_,
        "pos": token.pos_,
        # "tag": token.tag_,
        # "dep": token.dep_,
        # "shape": token.shape_,
        # "is_alpha": token.is_alpha,
        # "is_stop": token.is_stop
    })

print(tagged_words)
```

```python
# 1. Load existing JSON
with open(file_path, "r", encoding="utf-8") as f:
    story = json.load(f)

# 2. Add new field
story["tokenized_english_version"] = tagged_words

# 3. Save back to file
with open(file_path, "w", encoding="utf-8") as f:
    json.dump(story, f, indent=4, ensure_ascii=False)

print("Tokenized data added successfully!")
```
